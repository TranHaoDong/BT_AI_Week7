```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sas

url = 'Advertising.csv'
data= pd.read_csv(url)

data.head()
sas.pairplot(data)

```

```python
correlation = data.corr()
sas.heatmap(correlation, annot = True)
```

```python
data.plot(kind= 'scatter', x = 'TV', y = 'sales', figsize = (6 , 6), color = 'blue')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['TV']), y= data['sales'])
print(model.intercept_)
print(model.coef_)
```

```python
#Do chinh xac
model.score(X = pd.DataFrame(data['TV']), y= data['sales'])
```

```python
#Mo hinh bac 1
data.plot(kind= 'scatter', x = 'TV', y= 'sales', figsize = (5,5), color = 'r')
y_pred = model.predict(X = pd.DataFrame(data['TV']))
plt.plot(data['TV'], y_pred, color = 'blue')
     
```

```python
#Mo hinh bac 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['TV'], data['TV']**2]).T
model.fit(X, y= data['sales'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coefficient cua mo hinh', model.coef_)
print('Do chinh xac cua mo hinh', model.score(X, y = data['sales']))

#Ve do thi mo hinh bac 2
import numpy as np
range = np.arange(1.5,300,0.1)
data.plot(kind= 'scatter', x = 'TV', y= 'sales', figsize = (5,5), color = 'red')
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'green')

     
```
